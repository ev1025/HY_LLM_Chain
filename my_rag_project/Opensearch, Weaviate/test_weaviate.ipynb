```python
from dotenv import load_dotenv
load_dotenv()

import os, json
from utils.rag_utils_wv import get_weaviate_client

client = get_weaviate_client()
all_collections = client.collections.list_all()
    
# 딕셔너리에서 이름만 리스트로 추출
index_names = list(all_collections.keys())

print("\n--- Weaviate 인덱스 목록 ---")
print(index_names)
```

```python
# Weaviate 컬렉션 조회
COLLECTION_NAME = "Kgs" 
try:
    if not client.collections.exists(COLLECTION_NAME):
        print(f"🚨 '{COLLECTION_NAME}' 컬렉션이 존재하지 않습니다.")
    else:
        collection = client.collections.get(COLLECTION_NAME)
        
        response = collection.query.fetch_objects(
            limit=5,
            include_vector=False
            # return_properties 인자 생략 시 모든 속성 반환
        )
        
        print(f"\n--- '{COLLECTION_NAME}' 데이터 (최대 5개) ---")
        for item in response.objects:
            print(json.dumps(item.properties, indent=2, ensure_ascii=False))
            print("-" * 20)

except Exception as e:
    print(f"🚨 조회 중 오류: {e}")
```

```python
# 컬렉션 삭제 
COLLECTION_TO_DELETE = "Kgs" # 또는 "Kgs"
for COLLECTION_TO_DELETE in index_names:
    client.collections.delete(COLLECTION_TO_DELETE)
    print(f"✅ '{COLLECTION_TO_DELETE}' 컬렉션을 성공적으로 삭제했습니다.")
```
